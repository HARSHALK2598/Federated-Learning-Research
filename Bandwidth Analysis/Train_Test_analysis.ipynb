```python
# Function to extract data from logs
def extract_data(log):
    pattern = re.compile(
        r"Epoch (\d+)/50: Train Loss = ([\d.]+), Train Accuracy = ([\d.]+)%, Test Accuracy = ([\d.]+)%, Epoch Time = ([\d.]+)s, Total Time = ([\d.]+)s"
    )
    data = {
        "epoch": [],
        "train_loss": [],
        "train_accuracy": [],
        "test_accuracy": [],
        "epoch_time": [],
        "total_time": [],
    }
    for match in pattern.finditer(log):
        data["epoch"].append(int(match.group(1)))
        data["train_loss"].append(float(match.group(2)))
        data["train_accuracy"].append(float(match.group(3)))
        data["test_accuracy"].append(float(match.group(4)))
        data["epoch_time"].append(float(match.group(5)))
        data["total_time"].append(float(match.group(6)))
    return data
```

```python

# Extract data for all experiments
exp0_data = extract_data(exp0_log)
exp1_data = extract_data(exp1_log)
exp2_data = extract_data(exp2_log)
```

```python
# Plotting
def plot_metrics(exp0_data, exp1_data, exp2_data):
    plt.figure(figsize=(12, 8))

    # Train Accuracy vs Time
    plt.subplot(2, 2, 1)
    plt.plot(exp0_data["total_time"], exp0_data["train_accuracy"], label="Exp 0 (Baseline)", marker="o")
    plt.plot(exp1_data["total_time"], exp1_data["train_accuracy"], label="Exp 1 (FP16 + Compression)", marker="s")
    plt.plot(exp2_data["total_time"], exp2_data["train_accuracy"], label="Exp 2 (Compression)", marker="^")
    plt.xlabel("Total Time (s)")
    plt.ylabel("Train Accuracy (%)")
    plt.title("Train Accuracy vs Total Time")
    plt.legend()
    plt.grid(True)

    # Train Error vs Time
    plt.subplot(2, 2, 2)
    plt.plot(exp0_data["total_time"], [100 - acc for acc in exp0_data["train_accuracy"]], label="Exp 0 (Baseline)", marker="o")
    plt.plot(exp1_data["total_time"], [100 - acc for acc in exp1_data["train_accuracy"]], label="Exp 1 (FP16 + Compression)", marker="s")
    plt.plot(exp2_data["total_time"], [100 - acc for acc in exp2_data["train_accuracy"]], label="Exp 2", marker="^")
    plt.xlabel("Total Time (s)")
    plt.ylabel("Train Error (%)")
    plt.title("Train Error vs Total Time")
    plt.legend()
    plt.grid(True)

    # Test Accuracy vs Time
    plt.subplot(2, 2, 3)
    plt.plot(exp0_data["total_time"], exp0_data["test_accuracy"], label="Exp 0 (Baseline)", marker="o")
    plt.plot(exp1_data["total_time"], exp1_data["test_accuracy"], label="Exp 1 (FP16 + Compression)", marker="s")
    plt.plot(exp2_data["total_time"], exp2_data["test_accuracy"], label="Exp 2", marker="^")
    plt.xlabel("Total Time (s)")
    plt.ylabel("Test Accuracy (%)")
    plt.title("Test Accuracy vs Total Time")
    plt.legend()
    plt.grid(True)

    # Time Per Epoch
    plt.subplot(2, 2, 4)
    plt.plot(exp0_data["epoch"][1:], exp0_data["epoch_time"][1:], label="Exp 0 (Baseline)", marker="o")
    plt.plot(exp1_data["epoch"][1:], exp1_data["epoch_time"][1:], label="Exp 1 (FP16 + Compression)", marker="s")
    plt.plot(exp2_data["epoch"][1:], exp2_data["epoch_time"][1:], label="Exp 2", marker="^")
    plt.xlabel("Epoch")
    plt.ylabel("Time Per Epoch (s)")
    plt.title("Time Per Epoch")
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()
```

```python
# Call the plot function
plot_metrics(exp0_data, exp1_data, exp2_data)
```
